```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
mu = 0.5
sigma = 0.1

np.random.seed(0)

x = np.random.normal(mu, sigma, (400, 1))
y = np.random.normal(mu*2, sigma*3,(400, 1))
```

```python
plt.scatter(x, y, color='g', marker = 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
